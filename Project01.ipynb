## Show Data

Show some interesting findings.

```python
frequency = {}

for ad in adData:
    for item in ad["label_values"]:
        if item.get("title") == "Owner":
            owner_fields = item["dict"][0]["dict"]
            for field in owner_fields:
                if field.get("label") == "Name":
                    brand = field.get("value")
                    if brand in frequency:
                        frequency[brand] += 1
                    else:
                        frequency[brand] = 1

for brand, count in frequency.items():
    print(f"{brand}: {count}")
```
